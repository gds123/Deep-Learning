```python
import json
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_palette("Greens_r")

data = [json.loads(line)["overall"] for line in open("data/reviews_Amazon_Instant_Video_5.json", "rt")]
sns.distplot(data, kde=False, bins=5, axlabel="Score")
plt.ylabel("Number of Ratings")
plt.title("Distribution of Reviews")
plt.show()
# I guess you only rate movies you like? I'm surprised people don't rate movies they hate.... 
# maybe they never finish those movies in the first place, or just get sick of it and stop
```
